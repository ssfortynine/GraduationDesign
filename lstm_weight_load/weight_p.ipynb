```python
import torch  
import numpy as np  
import torch.nn as nn

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.lstm1 = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        # self.lstm2 = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        # self.lstm3 = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        # self.lstm4 = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):

        out, _ = self.lstm1(x) # shape（seq_Len, batch, num_directions * hidden_size)
        # out, _ = self.lstm2(out)
        # out, _ = self.lstm3(out)
        # out, _ = self.lstm4(out)

        out = self.fc(out[:, -1, :])  # 使用最后一个时间步的输出的输入（seq_Len, batch, num_directions * hidden_size)
        # print(out.shape)
        return out



model = LSTMModel(input_size=5, hidden_size=128, output_size=1, num_layers=1)
model.load_state_dict(torch.load('./seq20_last.pth'))
weight = model.state_dict()['lstm1.weight_ih_l0']


weight_i0 = model.state_dict()['lstm1.weight_ih_l0'][:128, :]
weight_i1 = model.state_dict()['lstm1.weight_hh_l0'][:128, :]
weight_ibias0 = model.state_dict()['lstm1.bias_ih_l0'][:128]

weight_ibias1 = model.state_dict()['lstm1.bias_hh_l0'][:128]
weight_f0 = model.state_dict()['lstm1.weight_ih_l0'][128:256, :]
weight_f1 = model.state_dict()['lstm1.weight_hh_l0'][128:256, :]
weight_fbias0 = model.state_dict()['lstm1.bias_ih_l0'][128:256]
weight_fbias1 = model.state_dict()['lstm1.bias_hh_l0'][128:256]
weight_c0 = model.state_dict()['lstm1.weight_ih_l0'][256:384, :]
weight_c1 = model.state_dict()['lstm1.weight_hh_l0'][256:384, :]
weight_cbias0 = model.state_dict()['lstm1.bias_ih_l0'][256:384]
weight_cbias1 = model.state_dict()['lstm1.bias_hh_l0'][256:384]
weight_o0 = model.state_dict()['lstm1.weight_ih_l0'][384:512, :]
weight_o1 = model.state_dict()['lstm1.weight_hh_l0'][384:512, :]
weight_obias0 = model.state_dict()['lstm1.bias_ih_l0'][384:512]
weight_obias1 = model.state_dict()['lstm1.bias_hh_l0'][384:512]
weight_class = model.state_dict()['fc.weight']
weight_class_bias = model.state_dict()['fc.bias']

weight_strings = []
weight_i_strings = [] 
weight_f_strings = []
weight_c_strings = [] 
weight_o_strings = []
weight_class_strings = []

 
   
weight = weight.numpy()
weight_i0 = weight_i0.numpy()  # 转换为 NumPy 数组  
weight_i1 = weight_i1.numpy()
weight_ibias = weight_ibias0.numpy() + weight_ibias1.numpy()

weight_f0 = weight_f0.numpy()  # 转换为 NumPy 数组  
weight_f1 = weight_f1.numpy()
weight_fbias = weight_fbias0.numpy() + weight_fbias1.numpy()

weight_c0 = weight_c0.numpy()  # 转换为 NumPy 数组  
weight_c1 = weight_c1.numpy()
weight_cbias = weight_cbias0.numpy() + weight_cbias1.numpy()

weight_o0 = weight_o0.numpy()  # 转换为 NumPy 数组  
weight_o1 = weight_o1.numpy()
weight_obias = weight_obias0.numpy() + weight_obias1.numpy()

weight_class = weight_class.numpy() 
weight_class_bias = weight_class_bias.numpy()

weight_i = np.zeros((128, 133))
weight_i[:, :5] = weight_i0
weight_i[:, 5:156] = weight_i1

weight_f = np.zeros((128, 133))
weight_f[:, :5] = weight_f0
weight_f[:, 5:156] = weight_f1

weight_c = np.zeros((128, 133))
weight_c[:, :5] = weight_c0
weight_c[:, 5:156] = weight_c1

weight_o = np.zeros((128, 133))
weight_o[:, :5] = weight_o0
weight_o[:, 5:156] = weight_o1
 

for vector in weight:  
    # 将向量转换为字符串列表，并格式化为带有 6 位小数的浮点数  
    format_vector = ['{:.6e}'.format(x) for x in vector]  
    # 将字符串列表转换为单个字符串，并用方括号包含起来  
    vector_str = '{' + ', '.join(format_vector) + '}'  
    # 将字符串添加到列表中  
    weight_strings.append(vector_str)  
  
# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight.txt', 'w') as f:  
    for vector_str in weight_strings:  
        f.write(vector_str + ',' + '\n')

for vector in weight_i:  
    # 将向量转换为字符串列表，并格式化为带有 6 位小数的浮点数  
    formatted_vector = ['{:.6e}'.format(x) for x in vector]  
    # 将字符串列表转换为单个字符串，并用方括号包含起来  
    weight_i_str = '{' + ', '.join(formatted_vector) + '}'  
    # 将字符串添加到列表中  
    weight_i_strings.append(weight_i_str)  
  
# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_i.txt', 'w') as f:  
    for weight_i_str in weight_i_strings:  
        f.write(weight_i_str + ',' + '\n')

for vector in weight_f:  
    # 将向量转换为字符串列表，并格式化为带有 6 位小数的浮点数  
    formatted_vector = ['{:.6e}'.format(x) for x in vector]  
    # 将字符串列表转换为单个字符串，并用方括号包含起来  
    weight_f_str = '{' + ', '.join(formatted_vector) + '}'  
    # 将字符串添加到列表中  
    weight_f_strings.append(weight_f_str)  
  
# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_f.txt', 'w') as f:  
    for weight_f_str in weight_f_strings:  
        f.write(weight_f_str + ',' + '\n')

for vector in weight_c:  
    # 将向量转换为字符串列表，并格式化为带有 6 位小数的浮点数  
    formatted_vector = ['{:.6e}'.format(x) for x in vector]  
    # 将字符串列表转换为单个字符串，并用方括号包含起来  
    weight_c_str = '{' + ', '.join(formatted_vector) + '}'  
    # 将字符串添加到列表中  
    weight_c_strings.append(weight_c_str)  
  
# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_c.txt', 'w') as f:  
    for weight_c_str in weight_c_strings:  
        f.write(weight_c_str + ',' + '\n')

for vector in weight_o:  
    # 将向量转换为字符串列表，并格式化为带有 6 位小数的浮点数  
    formatted_vector = ['{:.6e}'.format(x) for x in vector]  
    # 将字符串列表转换为单个字符串，并用方括号包含起来  
    weight_o_str = '{' + ', '.join(formatted_vector) + '}'  
    # 将字符串添加到列表中  
    weight_o_strings.append(weight_o_str)  
  
# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_o.txt', 'w') as f:  
    for weight_o_str in weight_o_strings:  
        f.write(weight_o_str + ',' + '\n')


 
formatted_vector = ['{:.6e}'.format(x) for x in weight_ibias]  
weight_ib_str  = ', '.join(formatted_vector)

# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_ib.txt', 'w') as f:  
    f.write(weight_ib_str + '\n')

formatted_vector = ['{:.6e}'.format(x) for x in weight_fbias]  
weight_fb_str  = ', '.join(formatted_vector)

# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_fb.txt', 'w') as f:  
    f.write(weight_fb_str + '\n')


formatted_vector = ['{:.6e}'.format(x) for x in weight_cbias]  
weight_cb_str  = ', '.join(formatted_vector)

# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_cb.txt', 'w') as f:  
    f.write(weight_cb_str + '\n')

formatted_vector = ['{:.6e}'.format(x) for x in weight_obias]  
weight_ob_str  = ', '.join(formatted_vector)

# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_ob.txt', 'w') as f:  
    f.write(weight_ob_str + '\n')

for vector in weight_class:  
    # 将向量转换为字符串列表，并格式化为带有 6 位小数的浮点数  
    formatted_vector = ['{:.6e}'.format(x) for x in vector]  
    # 将字符串列表转换为单个字符串，并用方括号包含起来  
    weight_class_str = '{' + ', '.join(formatted_vector) + '}'
    # 将字符串添加到列表中  
    weight_class_strings.append(weight_class_str)  
  
# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_class.txt', 'w') as f:  
    for weight_class_str in weight_class_strings:  
        f.write(weight_class_str + ',' + '\n')
        

formatted_vector = ['{:.6e}'.format(x) for x in weight_class_bias]  
weight_classb_str  = ', '.join(formatted_vector)

# 将所有向量的字符串写入文件，每个向量占一行  
with open('weight_classb.txt', 'w') as f:  
    f.write(weight_classb_str + '\n')


```
